```
import streamlit as st
import requests

# ==========================================
# MILESTONE 7: Frontend Development
# ==========================================

# Configure Streamlit page
st.set_page_config(
    page_title="Exam Anxiety Detector",
    page_icon="🧠",
    layout="centered"
)

# API Endpoint
API_URL = "http://localhost:8000/predict"

# Header and Introduction
st.title("🧠 AI-Based Exam Anxiety Detector")
st.markdown("""
Welcome to the Exam Anxiety Support System.
Express your pre-exam thoughts, feelings, or worries below. Our AI will analyze your text to gauge your stress levels and provide helpful, calming tips.
""")

# User Input area
user_text = st.text_area("How are you feeling about your upcoming exams?",
                         placeholder="e.g., I've been studying all week, but I'm still terrified I will forget everything during the test...",
                         height=150)

# Analyze Button
if st.button("Analyze Anxiety Level", type="primary"):
    if user_text.strip() == "":
        st.warning("Please enter some text to analyze.")
    else:
        with st.spinner("Analyzing your thoughts with BERT Model..."):
            try:
                # Send request to FastAPI backend
                response = requests.post(API_URL, json={"text": user_text})

                if response.status_code == 200:
                    result = response.json()
                    anxiety_level = result["anxiety_level"]
                    confidence = result["confidence"]

                    st.divider()

                    # Display Results based on Anxiety Category
                    if anxiety_level == "Low Anxiety":
                        st.success(f"**Result: {anxiety_level}** 🟢 (Confidence: {confidence:.2f})")
                        st.markdown("### ✨ Great job!")
                        st.info("**Tip:** You seem to be in a good headspace! Keep up the positive momentum. Remember to get a good night's sleep before the exam and trust your preparation.")

                    elif anxiety_level == "Moderate Anxiety":
                        st.warning(f"**Result: {anxiety_level}** 🟡 (Confidence: {confidence:.2f})")
                        st.markdown("### 🧘 Take a deep breath.")
                        st.info("**Tip:** It's completely normal to feel a bit nervous! Try the 4-7-8 breathing technique (inhale for 4s, hold for 7s, exhale for 8s). Break your study material into smaller, manageable chunks.")

                    elif anxiety_level == "High Anxiety":
                        st.error(f"**Result: {anxiety_level}** 🔴 (Confidence: {confidence:.2f})")
                        st.markdown("### ❤️ You are not alone.")
                        st.info("**Tip:** You are experiencing high stress. Please step away from your books for 15 minutes. Drink some water, take a short walk, or listen to calming music. If you feel overwhelmed, consider reaching out to a campus counselor or a trusted friend.")

                else:
                    st.error("Error connecting to the analysis engine. Please try again later.")
            except requests.exceptions.ConnectionError:
                st.error("Backend API is not reachable. Please ensure the FastAPI server is running on localhost:8000.")

# Footer with strict ethical disclaimers
st.divider()
st.caption("""
**Disclaimer:** This system is an AI-based supportive tool designed for educational mental wellness awareness.
It is **NOT** a diagnostic tool. Your data is handled anonymously. If you are experiencing severe distress, please contact a healthcare professional or your institution's mental health services immediately.
""")
```
